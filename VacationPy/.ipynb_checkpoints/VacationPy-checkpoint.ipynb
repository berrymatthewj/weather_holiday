```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
from pprint import pprint
from gkey import gkey
```

<font size="5">Store Part I Results into Dataframe</font> 

```python
#import data from part i
df=pd.read_csv(os.path.join("..","WeatherPy","output_data","clean_cities.csv"))
df.head()
```

```python
#drop weird new column
df1=df.drop(["Unnamed: 0"], axis=1)
df1.head()
```

<font size="5">Humidity Heatmap</font> 

```python
#configure gmaps
gmaps.configure(api_key=gkey)
locations_list=df1[["Latitude","Longitude"]].astype(float)
humidity_list=df1["Humidity"].astype(float)
humidity_list.max()
```

```python
#add heatmap layer
m = gmaps.figure(center=(39.754185, -105.230484), zoom_level=2, layout={'width': '800px','height': '600px',
'padding': '3px','border': '1px solid black'})
hm_layer=gmaps.heatmap_layer(locations_list,weights=humidity_list,dissipating=False,max_intensity=100,point_radius=5)
m.add_layer(hm_layer)
m
```

<font size="5"> Create New DataFrame Fitting Weather Criteria</font> 

```python
#narrow down based on my weather preferences
hotel_df=df1[(df1["Maximum Temperature"]>=80)&(df1["Maximum Temperature"]<=90)&(df1["Wind Speed"]<10)&(df1["Cloudiness"]<=10)]
hotel_df
```

<font size="5">Hotel Map</font> 

```python
#create an API query
base_url="https://maps.googleapis.com/maps/api/place/nearbysearch/json?"
#empty list for data from query
hotel_name=[]

#run query and collect first hotel within 5000m of the cities in the hotel_df
for index, row in hotel_df.iterrows():
    latitude=row["Latitude"]
    longitude=row["Longitude"]
    city=row["City"]
    print(f"Retrieving Results for Index {index}: {city}.")
    response=requests.get(f"{base_url}location={latitude},{longitude}&radius=5000&type=lodging&key={gkey}").json()
    results=response["results"]
    try:
        hotel_name.append(results[0]["name"])
        print(f"The closest place to stay is {results[0]['name']}.")
    except Exception:
        print(f"Error")
        hotel_name.append("NA")
    print("----------------------------------------")
```

```python
#add hotel name column
hotel_df["Hotel Name"]=hotel_name
hotel_df
```

```python
# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
marker_locations = hotel_df[["Latitude", "Longitude"]]
marker_layer = gmaps.marker_layer(marker_locations, info_box_content=hotel_info)
m.add_layer(marker_layer)
#display map with both layers
m
```
